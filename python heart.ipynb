```python
import turtle

t=turtle.Turtle()

t.color("red")
t.begin_fill()
t.fillcolor("red")
t.left(140)
t.forward(180)
t.circle(-90,200)

t.setheading(60)
t.circle(-90,200)
t.forward(180)
t.end_fill()

t.penup()

t.goto(-130,130)
t.pendown()
t.color('white')
t.write("love u ",font=("verdana",25,"bold"))
turtle.done()
```
